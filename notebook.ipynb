## Business idea: 
#### website EDUMENTOR. The clients fill the profile with the name of the desired country of education, level of education, their area of interest, financial capabilities, and so on. In addition, the website has a partnership with the language courses and offers the clients the possibility to take courses and receive the certificate. 

#### Variables:
##### name = name of the client
#####  self.client_ID = client ID
##### country_origins = name of the country of origins
##### country_desire = name of the country the client want to study in
##### education_level = level of desired education
##### language_spoken = the client's language spoken
##### language_education = the language of desired education
##### interest_area = subject or area of desired education
##### finance_personal = how much willing to spend on education per  year
##### education_type = paid or free education needed
##### language_consultation_needed = if client want an education in different language suggest the consultation
##### consultation_number = number of consultations offered
##### consultation_price = set by the mentor 
##### portfolio_made = the portfolio preparation (includes CV, motivation letter, etc)
##### portfolio_price = set by the mentor
##### income = how much did client spend
